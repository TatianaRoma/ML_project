```python
### libraries
import numpy as np
import pandas as pd
import matplotlib as mpl

import matplotlib.pyplot as plt
%matplotlib inline

from pylab import *
import itertools
import random
import scipy
import six
import re

from __future__ import print_function
from pprint import pprint
import time as time
import sys
import seaborn as sns; sns.set()

from time import time
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

from sklearn.svm import SVC 
from sklearn.naive_bayes import MultinomialNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

from sklearn.grid_search import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.cross_validation import KFold
from sklearn.metrics import confusion_matrix, f1_score
from sklearn import metrics
from sklearn.cross_validation import train_test_split
```

```python
path = r'/home/ubuntu/yelp/reviews_all.csv'
reviews = pd.read_csv(path)
reviews['Rating_nom'] = ["one" if Rating == 1 else "two" if Rating == 2 else "three" if Rating == 3 
                else "four" if Rating == 4 else "five" for Rating in reviews.Rating]
reviews = reviews[['Text','Rating_nom']]
reviews.head()
```

```python
labels = ["one","two","three","four","five"]

def plot_confusion(confmat):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    cax = ax.matshow(confmat)
    plt.title('Confusion matrix of the classifier')
    fig.colorbar(cax)
    ax.set_xticklabels([''] + labels)
    ax.set_yticklabels([''] + labels)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.show()
```

```python
rev_num = [10000, 20000, 30000, 40000, 50000, 60000, 70000, 80000, 90000, 100000]
```

```python
pipeline = Pipeline([
    ('vectorizer',  CountVectorizer(stop_words='english',
                                   max_df=0.7, min_df=100)),
    #('transformer', TfidfTransformer()),
    ('classifier',  MultinomialNB()) ])
nb_times = []
nb_scores = []
for num in rev_num:
    t0 = time()
    reviews_sample = reviews.sample(n = num)
    X_train, X_test, y_train, y_test = train_test_split(reviews_sample.Text.values, 
                                                        reviews_sample.Rating_nom.values, 
                                                        test_size=0.33)
    pipeline.fit(X_train, y_train)
    predictions = pipeline.predict(X_test)
        
    nb_times.append(time()-t0)
    nb_scores.append(f1_score(y_test, predictions, labels = labels, average = 'macro'))
```

```python
plt.plot(rev_num, nb_times)
plt.xlabel('Number of reviews in the sample')
plt.ylabel('Time to train and test 1 Naive Bayes model (sec)')
plt.title('Time it takes to fit a Naive Bayes model')
plt.show()
```

```python
pipeline = Pipeline([
    ('vectorizer',  CountVectorizer(ngram_range=(1,3), stop_words='english',
                                   max_df=0.7, min_df=100)),
    ('transformer', TfidfTransformer()),
    ('classifier',  LogisticRegression(multi_class='multinomial',solver='lbfgs')) ])
lr_times = []
lr_scores = []
for num in rev_num:
    t0 = time()
    reviews_sample = reviews.sample(n = num)
    X_train, X_test, y_train, y_test = train_test_split(reviews_sample.Text.values, 
                                                        reviews_sample.Rating_nom.values, 
                                                        test_size=0.33)
    pipeline.fit(X_train, y_train)
    predictions = pipeline.predict(X_test)
        
    lr_times.append(time()-t0)
    lr_scores.append(f1_score(y_test, predictions, labels = labels, average = 'macro'))
```

```python
plt.plot(rev_num, lr_times)
plt.xlabel('Number of reviews in the sample')
plt.ylabel('Time to train and test 1 Logistic Regression model (sec)')
plt.title('Time it takes to fit a Logistic Regression model')
plt.show()
```

```python
pipeline = Pipeline([
    ('vectorizer',  CountVectorizer(ngram_range=(1,3), stop_words='english',
                                   max_df=0.7, min_df=100)),
    ('transformer', TfidfTransformer()),
    ('classifier',  SVC(kernel='linear')) ])
svm_times = []
svm_scores = []
for num in rev_num:
    t0 = time()
    reviews_sample = reviews.sample(n = num)
    X_train, X_test, y_train, y_test = train_test_split(reviews_sample.Text.values, 
                                                        reviews_sample.Rating_nom.values, 
                                                        test_size=0.33)
    pipeline.fit(X_train, y_train)
    predictions = pipeline.predict(X_test)
        
    svm_times.append(time()-t0)
    svm_scores.append(f1_score(y_test, predictions, labels = labels, average = 'macro'))
```

```python
plt.plot(rev_num, svm_times)
plt.xlabel('Number of reviews in the sample')
plt.ylabel('Time to train and test 1 SVM model (sec)')
plt.title('Time it takes to fit an SVM model')
plt.show()
```

```python
plt.plot(rev_num, nb_scores)
plt.xlabel('Number of reviews in the sample')
plt.ylabel('F1 Score average')
plt.title('Naive Bayes: Accuracy')
plt.show()
```

```python
plt.plot(rev_num, lr_scores)
plt.xlabel('Number of reviews in the sample')
plt.ylabel('F1 Score average')
plt.title('Logistic Regression: Accuracy')
plt.show()
```

```python
plt.plot(rev_num, svm_scores)
plt.xlabel('Number of reviews in the sample')
plt.ylabel('F1 Score average')
plt.title('Support Vector Machines: Accuracy')
plt.show()
```
